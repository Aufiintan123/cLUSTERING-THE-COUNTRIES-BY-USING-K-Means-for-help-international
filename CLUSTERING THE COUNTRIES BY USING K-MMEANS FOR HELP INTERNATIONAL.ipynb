## Outliers Treatment

```python
plt.figure(figsize = (15, 15))
for index, column in enumerate(df.describe().columns):
    plt.subplot(3, 3, index + 1)
    sns.boxplot(x = column, data = df)
plt.show()
```

```python
columns_numeric = ['Kematian_anak', 'Ekspor', 'Kesehatan', 'Impor', 'Pendapatan', 'Inflasi', 'Harapan_hidup', 'Jumlah_fertiliti', 'GDPperkapita']
```

```python
Q1 = df[columns_numeric].quantile(0.25)
Q3 = df[columns_numeric].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - IQR * 1.5
upper = Q3 + IQR * 1.5
```

```python
df_non_outlier = df.copy()
for x in columns_numeric:
    filter_min = df_non_outlier[x] < lower[x]
    filter_max = df_non_outlier[x] > upper[x]
    df_non_outlier = df_non_outlier[~(filter_min | filter_max)]
```
